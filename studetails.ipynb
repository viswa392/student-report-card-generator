```python
student_id = input("Enter student id = ")
student_name = input("Enter student name = ")
```

```python
Tamil = int(input("Enter Tamil mark: "))
English = int(input("Enter English mark: "))
Maths = int(input("Enter Maths mark: "))
Science = int(input("Enter Science mark: "))
Computer = int(input("Enter Computer mark: "))
```

```python
student = {
    "student id": student_id,
    "student name": student_name,
    "mark": {
        "Tamil mark": Tamil,
        "English mark": English,
        "Maths mark": Maths,
        "Science mark": Science,
        "Computer mark": Computer
    }
}
print(student)
```

```python
print("\nSubject Marks")
print("Tamil :", student["mark"]["Tamil mark"])
print("English :", student["mark"]["English mark"])
print("Maths :", student["mark"]["Maths mark"])
print("Science :", student["mark"]["Science mark"])
print("Computer :", student["mark"]["Computer mark"])
```

```python
total = (
    student["mark"]["Tamil mark"] +
    student["mark"]["English mark"] +
    student["mark"]["Maths mark"] +
    student["mark"]["Science mark"] +
    student["mark"]["Computer mark"]
)

print("Total Marks :", total)
```

```python
total = sum(student["mark"].values())

average = round(total / len(student["mark"]), 2)

student["total"] = total
student["average"] = average

print("Total Marks :", student["total"])
print("Average Marks :", student["average"])
```
